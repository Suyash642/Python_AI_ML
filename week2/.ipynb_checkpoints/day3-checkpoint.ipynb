```python
serA= pd.Series([1,1,2,2,4],dtype=float)
serB = pd.Series([0.5,1.5,2.5,3.5,4.5])
print(f'Series 1 : \n{serA}')
print(f'\nSeries 2 : \n{serB}')
```

```python
diff_arr = np.subtract(serA,serB)
print(f'\nDiffrenced array is : \n{diff_arr}')
```

```python
squared_diff_arr = np.square(diff_arr)
print(f'\nSquare of differnced array is : \n{squared_diff_arr}')
```

```python
Mean_squared_diff_arr = np.mean(squared_diff_arr)
print(f'\nMean Squared Error is : \n{Mean_squared_diff_arr}')
```

```python
words=['science','maths','geography','physics','chemistry','biology','marathi','hindi']
words_ser = pd.Series(words)
print(f'\n Words series: \n{words_ser}')
```

```python
for key,word in words_ser.iteritems():
    #print(word)
    result = ' '
    result += word[:-1]+word[-1].upper()
    print(result)
```

```python
#importing diabetes dataset
data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv')
data.head()
```

```python
#replace 0 with NaN values
data = data.replace(to_replace = 0, value = np.nan)
data
```

```python
#replace NaN with mean of respective col
data.replace(to_replace = np.nan, value = data.mean())
```

```python
#mean of respective col
data.mean()
```

```python
#replace NaN with median of respective col
data.replace(to_replace = np.nan, value = data.median())
```

```python
#median of respective col
data.median()
```

```python
data.to_csv('Diabetes.csv')
```

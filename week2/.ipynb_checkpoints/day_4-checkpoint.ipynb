```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
#creating x
x = np.linspace(0,10,10)
x
```

```python
# for f(x) = e^(-x/10)
y1 = math.e**(-x/10)
y1
```

```python
#plotting graph
plt.plot(x,y1)
plt.title('Functional Plot_1')
plt.xlabel('X')
plt.ylabel('Y= e^(-x/10)')
```

```python
#for sin(πx) 
y2 = np.sin(math.pi*x)
y2
```

```python
#plot2
plt.plot(x,y2)
plt.title('Functional Plot_2')
plt.xlabel('X')
plt.ylabel('Y = sin(πx)')
```

```python
# g(x)=x*e^(-x/3)
y3 = x*math.e**(-x/3)
y3
```

```python
#plot3
plt.plot(x,y3)
plt.title('Functional Plot_3')
plt.xlabel('X')
plt.ylabel('Y = x^e(-x/3)')
```

```python
#multiple figure
fig = plt.figure(figsize=(5,4), dpi=100,facecolor="green")
axes_1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('X')
axes_1.set_ylabel('Y')
axes_1.set_title('Functional Plots')
axes_1.plot(x,y1,label='X/e^(-x/10)')
axes_1.plot(x,y2,label='X/ sin(πx) ')
axes_1.plot(x,y3 ,label='X/x^e(-x/3)')

axes_1.legend(loc=0)
```

```python
fig.savefig('Functional_plots.jpg')
```
